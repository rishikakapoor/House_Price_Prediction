## Categorical Variable Missing Value

Variables with more than 50% null values are labeled as other.

variables with less than 50% null values, the null values are replaced with the most frequent value.

```python
variables= ['Alley', "FireplaceQu","PoolQC","MiscFeature","Fence"]
for feature in variables:
    house_train_dataset[feature]=house_train_dataset[feature].fillna('other')
```

```python
other_variables= ['LotFrontage', 'MasVnrType', 'MasVnrArea', 'GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                  'BsmtFinType2', 'Electrical']
```

```python
for feature in other_variables:
    mode= house_train_dataset[feature].mode()
    house_train_dataset[feature]= house_train_dataset[feature].fillna(mode[0])
    
```

```python
#Lets create a heatmap to see which all columns has null values
plt.figure(figsize=(30,12))
sns.heatmap(house_train_dataset.isnull(), yticklabels=False, cmap='viridis',cbar='cyan')
```

## Encoding Categorical Variable 

```python
#encoding:
def encode_label(df, col):
    #encode the categories using average SalePrice for each category to replace label
    cat_dict ={}
    cats = df[col].cat.categories.tolist()
    for cat in cats:
        cat_dict[cat] = house_train_dataset[house_train_dataset[col]] == cat['SalePrice'].mean()  
    df[col] = df[col].map(cat_dict)
```

```python
for col in house_train_dataset.columns:
    if house_train_dataset[col].dtype.name == "category":
        encode_label(house_train_dataset, col) 
```
